```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 파일 불러오기
df = pd.read_csv('/content/title.basics.tsv', sep='\t', na_values='\\N')

# 데이터 전처리: 영화만 필터링
df = df[df['titleType'] == 'movie']
df['startYear'] = pd.to_numeric(df['startYear'], errors='coerce')
df.dropna(subset=['startYear', 'genres'], inplace=True)

# 장르 분리
genres = df['genres'].str.split(',').explode()

# Chart 1: 장르별 영화 수
plt.figure(figsize=(12, 6))
sns.countplot(x=genres, order=genres.value_counts().index)
plt.xticks(rotation=45)
plt.title('Number of movies by genre')
plt.xlabel('Genre')
plt.ylabel('Number of movies')
plt.show()

# Chart 2: 연도별 영화 제작 추이
plt.figure(figsize=(12, 6))
sns.histplot(df['startYear'], bins=30)
plt.title('Number of movies productions by year')
plt.xlabel('Year')
plt.ylabel('Number of movies')
plt.show()
```
